```python
outliers=[]
```

```python
Outliers = []
for col in df.columns:
    column_data = df[col]
    Q1 = column_data.quantile(0.25)
    Q3 = column_data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = column_data[(column_data < lower_bound) | (column_data > upper_bound)]
    Outliers.extend(outliers)
```

```python
Outliers
```

```python
# Convert outliers list to a set for faster lookup
outliers_set = set(Outliers)

# Print values which are not outliers
for col in df.columns:
    column_data = df[col]
    non_outliers = [value for value in column_data if value not in outliers_set]
    print("Non-outliers in column '{}':".format(col), non_outliers)

```
